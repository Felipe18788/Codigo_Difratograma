```python
#Importando o arquivo (Excel) com os dados experimentais da análise de DRX (Intensidade x Ângulo de difração)
import pandas as pd

data = pd.read_excel("Sample1.xlsx")

sample_angle = data["2Theta"]
sample_int = data["Int"]
```

```python
#Importando o arquivo (CSV) com os dados da referência do programa Profex
sample_phase = pd.read_csv("gibbsita_amostra1.csv", sep=' ')

#Tratando os dados do arquivo da referência 
sample_phase = sample_phase.drop(columns=["File","Phase","d (nm)","Rel. intensity (%)","Texture","B1 (1/nm)","B2 (1/nm)"])

phase_angle = sample_phase['Angle']
phase_int = sample_phase['Intensity (deg*cts)']
```

```python
#Montagem do difratograma com os picos da referência
import plotly.graph_objects as go

fig = go.Figure()
fig.add_trace(go.Scatter(
    x=sample_angle, 
    y=sample_int, 
    mode='lines',  
    name='Teste',
    line=dict(shape='spline', color='blue', width=1.4)
))

fig.update_layout(
    title=dict(text="Diffractogram", x=0.5, xanchor="center", y=0.85, font=dict(size=20, family="Times")),
    xaxis_title="Diffraction angle (2θ)",
    yaxis_title="Intensity",
    xaxis=dict(title_font=dict(size=16, family="Times")),  
    yaxis=dict(title_font=dict(size=16, family="Times")),
    legend=dict(x=1.05, y=1, xanchor="left", yanchor="top", font=dict(size=14, family="Times")),
)

for linha in range(len(sample_phase)):
    fig.add_shape(
        type='line',
        x0=sample_phase.iloc[linha]["Angle"], 
        x1=sample_phase.iloc[linha]["Angle"], 
        y0=0, 
        y1=sample_phase.iloc[linha]["Intensity (deg*cts)"],
        line=dict(color='red', width=1.3))
    

import numpy as np 
y_max = np.max(data["Int"])

fig.update_layout(
    xaxis=dict(range=[sample_angle[0], sample_angle[len(data)-1]]),  
    yaxis=dict(range=[0,y_max + y_max*0.1]),    
)

fig.update_layout(
    xaxis=dict(showline=True,  linecolor='black', linewidth=1.1, mirror=True),
    yaxis=dict(showline=True, linecolor='black', linewidth=1.1, mirror=True),
    plot_bgcolor='white'
)

fig.update_layout(
    xaxis=dict(showgrid=True, gridcolor='lightgray'),
    yaxis=dict(showgrid=True, gridcolor='lightgray')
)


fig.show()
```
